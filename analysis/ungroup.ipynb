```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


# Read the file
# this data contains 5.5E10 photons simulated with spekpy x-ray spectrum
df = pd.read_csv('data.csv', header=None)

# Correct the energy data
df[0] = np.arange(0, 40.1, 0.1)

# Define the plotting function
def plot_spectrum(df, column, title):
    sns.set_style("whitegrid")
    plt.rcParams['figure.dpi'] = 120 # set the dpi
    plt.rcParams.update({'font.size': 16}) # set the font size
    plt.figure(figsize=(8, 8))
    # plt.grid(color='white', linestyle='-', linewidth=2)
    plt.plot(df[0][1:], df[column][1:])  # Ignore the first data point
    plt.title('Energy Spectrum Deposited in {}'.format(title))
    plt.xlabel('X-ray Photon Energy (keV)')
    plt.ylabel('Number of Photons')
    plt.text(0.5, 0.9, 'Total Energy: {:.2f} keV'.format(df[column][1:].sum()), transform=plt.gca().transAxes)
    plt.show()

# Plot the graphs
plot_spectrum(df, 1, 'Lead Wall')
plot_spectrum(df, 2, 'Human1')
plot_spectrum(df, 3, 'Human2')
```
